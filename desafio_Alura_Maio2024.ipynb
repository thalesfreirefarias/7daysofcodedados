```python
dados_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
```

```python
#uni dois df
emprestimos_completo = emprestimos_biblioteca.merge(dados_exemplares)
emprestimos_completo
```

```python
# Garantir que a coluna 'data_emprestimo' seja do tipo datetime
emprestimos_completo['data_emprestimo'] = pd.to_datetime(emprestimos_completo['data_emprestimo'])

# Criar o data frame 'meses' com o formato desejado
emprestimos_completo['meses'] = emprestimos_completo['data_emprestimo'].dt.strftime('%Y-%b')

# Ordenar o DataFrame pela coluna 'data_emprestimo'
emprestimos_completo = emprestimos_completo.sort_values('data_emprestimo')

# Contar o número de ocorrências por ano
month_counts = emprestimos_completo['data_emprestimo'].dt.to_period('Y').value_counts().sort_index()

# Converter a Series para DataFrame
month_counts_df = month_counts.reset_index()
month_counts_df.index = month_counts_df.index + 1
month_counts.index.name = "Quantidade Emprestimo"
month_counts_df.columns = ['Anos', 'Quantidade de emprestimos']

# Converter a coluna 'Quantidade de emprestimos' para numérico
month_counts_df['Quantidade de emprestimos'] = pd.to_numeric(month_counts_df['Quantidade de emprestimos'])

# Aplicar o estilo
estilo_meses = month_counts_df.style
estilo_meses.format({'Quantidade de emprestimos': '{:,.0f}'}) \
          .highlight_max(color='green', subset=['Quantidade de emprestimos']) \
          .highlight_min(color='#F16165', subset=['Quantidade de emprestimos'])

estilo_meses
```

```python
# Salvar o DataFrame estilizado como HTML
estilo_meses.to_html('emprestimos_por_ano.html', escape=False)
```

```python
# Agrupar os dados por 'biblioteca' e contar a quantidade por biblioteca
uso_Biblioteca = emprestimos_completo['biblioteca'].value_counts()

uso_Biblioteca_df = uso_Biblioteca.reset_index()
uso_Biblioteca_df.index = uso_Biblioteca_df.index + 1

uso_Biblioteca_df.columns = ['Biblioteca', 'Quantidade de Empréstimos']

# Aplicar o estilo ao DataFrame, especificando a coluna de números
estilo_Biblioteca = uso_Biblioteca_df.style.format({'Quantidade de Empréstimos': '{:,.0f}'}) \
                                           .highlight_max(subset='Quantidade de Empréstimos', color='green') \
                                           .highlight_min(subset='Quantidade de Empréstimos', color='#F16165')


# Exibindo o DataFrame estilizado
estilo_Biblioteca
```

```python
#  Se for menor ou maior criar uma lista com outro nome
CDU_lista = []
for CDU in emprestimos_completo['localizacao']:
  if(CDU < 100):
    CDU_lista.append('Generalidades')
  elif(CDU < 200):
    CDU_lista.append('Filosofia e psicologia')
  elif(CDU < 300):
    CDU_lista.append('Religião')
  elif(CDU < 400):
    CDU_lista.append('Ciências sociais')
  elif(CDU < 500):
    CDU_lista.append('Classe vaga')
  elif(CDU < 600):
    CDU_lista.append('Matemática e ciências naturais')
  elif(CDU < 700):
    CDU_lista.append('Ciências aplicadas')
  elif(CDU < 800):
    CDU_lista.append('Belas artes')
  elif(CDU < 900):
    CDU_lista.append('Linguagem')
  else:
    CDU_lista.append('Geografia. Biografia. História.')

emprestimos_completo['CDU_geral'] = CDU_lista
```

```python
#Quantidade de exemplares emprestados por ano
emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data.index = emprestimos_data.index + 1
emprestimos_data
```

```python
emprestimos_por_ano = emprestimos_data.groupby(emprestimos_data['data'].dt.year)['quantidade'].sum()
emprestimos_data.columns = ['Ano','quantidade']
emprestimos_por_ano
```

```python
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='deep',
              font_scale=1.3,
              rc={"figure.figsize":(15,8)})
```

```python
# Criar o gráfico de linha
ax = sns.lineplot(data=emprestimos_por_ano, x='Ano', y='quantidade')

# Remover os rótulos dos eixos x e y
ax.set(xlabel=None, ylabel=None)

# Rotacionar os rótulos do eixo x para melhor legibilidade
ax.tick_params(axis='x', rotation=30)

# Formatar os rótulos do eixo y para usar pontos como separadores de milhar
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))

# Definir o título do gráfico
ax.set_title('Quantidade de exemplares emprestados do SISBI por ano' + '\n', size=20, loc='left', weight='bold')

# Mostrar o gráfico
plt.show()
```

```python
emprestimos_por_mes = emprestimos_data.groupby(by=emprestimos_data.data.dt.month).sum()
emprestimos_por_mes.index.name = 'mes'
emprestimos_por_mes
```

```python
#Modificando os números por nomes dos meses

dicionario_meses = {1:'Jan',2:'Fev',3:'Mar',4:'Abr',
                    5:'Mai',6:'Jun',7:'Jul',8:'Ago',
                    9:'Set',10:'Out',11:'Nov',12:'Dez'}


emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)


emprestimos_por_mes
```

```python
ax = sns.lineplot(data=emprestimos_por_mes,x='mes',y='quantidade')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x',rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title("Quantidade de exemplares emprestados do SISBI por mês"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=265000,fontsize=18, ha='left',color='gray')

ax=ax
```
